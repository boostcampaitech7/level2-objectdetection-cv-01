```python
from diffusers import StableDiffusionUpscalePipeline
from PIL import Image
import torch

# 모델 불러오기
model_id = "stabilityai/stable-diffusion-x4-upscaler"
pipeline = StableDiffusionUpscalePipeline.from_pretrained(model_id)
pipeline = pipeline.to("cuda")  # GPU 사용 가능 시

# 이미지를 로드하여 변환하기 (로컬 파일 사용 가능)
image = Image.open("/home/taeyoung4060ti/바탕화면/level2-objectdetection-cv-01/ultralytics-main/dataset/train/0004.jpg")

# 업스케일링 수행
upscaled_image = pipeline(image=image, prompt="make to high quality image").images[0]

# 결과 저장
upscaled_image.save("upscaled_image.jpg")
```
